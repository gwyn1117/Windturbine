```python
df['target']=df['功率']/df['理论应发功率']
```

```python
df['target']=np.round(df['target'],1)
```

```python
def simple_features(sub_df,length=5):
    row_list=[i for i in sub_df.describe().iterrows()][1:]
    kv_list=[]
    for row in row_list:
        kv_list+=[(row[0]+"_"+col[0],col[1]) for col in row[1].iteritems()][:-1]
    previous_target=[("Previous_target_"+i[0],i[1]) for i in sub_df.iloc[:-1]['target'].describe().iteritems()][1:]
    kv_list+=previous_target
    kv_list.append(("target",sub_df['target'].values[-1]))
    kv_list.append(("time",sub_df['时间'].values[0]))
    feaetures=pd.DataFrame(kv_list).set_index(0).T
    #feaetures['target'][1]=sub_df['target'].values[-1]
    return feaetures
```

```python
feature_list=[]
for i in range(df.shape[0]-5):
    feature_list.append(simple_features(df.iloc[i:i+5],length=5))
```

```python
all_feature_df=pd.concat(feature_list,axis=0)
```

```python
all_feature_df.reset_index(drop=True).to_csv("test_feaetures.csv")
```

```python

feature_filled=feature_df.copy()
#feature_filled['target']=np.log(feature_filled['target'])
feature_filled=feature_filled.replace(np.nan,0).replace(np.inf,9999).replace(-np.inf,-9999)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
regr = RandomForestRegressor(max_depth=10, random_state=0,n_estimators=100,n_jobs=-1)
```

```python
regr.fit(feature_filled.drop(columns=['target','time']),feature_filled['target'])
```

```python
regr.score(feature_filled.drop(columns=['target','time']),feature_filled['target'])
```

```python
from sklearn.tree import export_graphviz
# Export as dot file
export_graphviz(regr.estimators_[1], out_file='tree.dot', 
                feature_names = feature_filled.columns[:-2],
#                 class_names = iris.target_names,
                rounded = True, proportion = False, 
                precision = 2, filled = True)

# Convert to png using system command (requires Graphviz)
from subprocess import call
call(['dot', '-Tpng', 'tree.dot', '-o', 'tree.png', '-Gdpi=600'])

# Display in jupyter notebook
from IPython.display import Image
Image(filename = 'tree.png')
```
